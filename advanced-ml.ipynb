### Almost all of the fraud transactions is less than 2500.
Lets calculate the percentage and check if it is true.

```python
# Calculate that number of transactions the are either below 2500 or above

new_data = data[['Time','Amount', 'Class']]

amount_above = 0
amount_below = 0

for i in range(new_data.shape[0]):
    if (new_data.iloc[i]['Amount'] < 2500):
        amount_below += 1
    else:
        amount_above += 1

print("Number of transactions that exceeds 2500:", amount_above)
print('Number of transactions that are below 2500:', amount_below) 
```

```python
percentage = (amount_below/data.shape[0])*100

print('Percentage of transactions that are less than 2500:', percentage)
```

```python
valid = 0
fraud = 0

for i in range(new_data.shape[0]):
    if (new_data.iloc[i]['Amount'] < 2500):
        if (new_data.iloc[i]['Class'] == 0):
            valid += 1
        else:
            fraud += 1

print('Number of fraud cases:', fraud)
print('Number of valid cases:', valid)  
```
