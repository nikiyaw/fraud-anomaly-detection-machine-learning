## Let study the relationship of fraud transactions with respect to the amount of money

```python
# Plot and compare different features side-by-side

def pairplot(data, feature_1, feature_2, target):
    sns.FacetGrid(data, hue=target, height=6).map(plt.scatter, feature_1, feature_2).add_legend()
    plt.show()
```

```python
pairplot(data, 'Time', 'Amount', 'Class')
```

### Almost all of the fraud transactions is less than 2500.

```python
# Calculate that number of transactions the are either below 2500 or above

new_data = data[['Time','Amount', 'Class']]

amount_above = 0
amount_below = 0

for i in range(new_data.shape[0]):
    if (new_data.iloc[i]['Amount'] < 2500):
        amount_below += 1
    else:
        amount_above += 1

print("Number of transactions that exceeds 2500:", amount_above)
print('Number of transactions that are below 2500:', amount_below) 
```
